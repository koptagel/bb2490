```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
filename = "../data/supplementary_data/gene_group_log2.xls"
xls = pd.read_excel(filename)

print(xls.shape)
xls.head()
```

```python
selected_columns = xls.columns[1:11]
print(selected_columns)
mat = xls.as_matrix(columns=selected_columns)
print(mat.shape)
print(np.min(mat), np.max(mat))
```

```python
plt.figure(figsize=(20,10))
plt.imshow(mat, interpolation="nearest", aspect="auto", cmap=cm.coolwarm, vmin=-3, vmax=3)
plt.xticks(np.arange(mat.shape[1]),["U_R1","U_R2","U_R3","U_R4","P_R1","P_R2","P_R3","M_R1","M_R2","M_R3"])
plt.xlabel("Replicates")
plt.ylabel("Gene Symbols")
plt.title("Corrected log2 ratios (min: %.2f, max: %.2f)" % (np.min(mat), np.max(mat)))
plt.colorbar()
plt.show()
```
